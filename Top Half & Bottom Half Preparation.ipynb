### Top-half part and Bottom-half part preparation

######  Top-half part preparation

```python
import numpy as np
x_3000_train = np.load('3000_x.npy')
print(x_3000_train.shape)
```

```python
for j in range(len(x_3000_train)):
    for i in range(14,28):
        x_3000_train[j][i] = 0
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for i in range(15):
    plt.imshow(x_3000_train[i],cmap = 'gray')
    plt.show()
    
```

```python
print(x_3000_train.shape)
```

```python
np.save('x_3000_topview',x_3000_train)
```

###### Bottom-half part preparation
